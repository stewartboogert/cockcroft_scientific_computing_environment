## Package manager

The easiest way to install scientifc software is via a package manager. Almost all modern computing systems have a package manager, that downloads, installs, updates and uninstalls software. The are two main package managers for Mac (homebrew and macports). 

1. Open Terminal
1. Install homebrew
2. Exit terminal and restart
3. Test brew

### Getting help on homebrew commands 

```python
%%bash
brew help
```

### Listing installed packages

```python
%%bash
brew list
```

### Discovering/searching for packages 

```python
%%bash
brew search hdf5
```

### Installing a package

```python
%%bash 
brew install hdf5
```
