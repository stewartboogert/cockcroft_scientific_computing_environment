## Why package managers?

The easiest way to install scientifc software is via a package manager. Almost all modern computing systems have a package manager, that downloads, installs, updates and uninstalls software. There are many good reasons to use a package manager for your scientific software stack. 

1. Discovering packages 
2. Repoducibility
3. Installing depencencies
4. Handing conflicts (some packages cannot be installed together)
5. Uninstalling cleanly
6. Avoids custom compilation and/or installation
