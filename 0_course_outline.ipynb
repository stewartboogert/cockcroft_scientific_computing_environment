### Software stack

Scientific software is a complex ecosystem which is often complex and difficult to maintain. Most of us are just interested in doing science, to achieve this we need a stable and repoducible software environment that we can trust. Here is an example of a python focused stack  

| Depth  | Stack layer               | Examples                               | Lecture |
|--------|---------------------------|----------------------------------------|---------| 
| Top    | Your Research Code        | notebooks, scripts, analysis           | 4       |
|        | Visualisation & Reporting | Matplotlib, Plotly, Jupyter            | 3       |
|        | Scientific Libraries      | NumPy, SciPy, Pandas, Sklearn, pytorch | 3       |
|        | Language                  | Python                                 | 3       |
|        | Environment Manager       | Conda / virtualenv                     | 2       |
|        | Python package Manager    | pip / conda                            | 2       |
|        | Package Manager           | apt/yum/brew                           | 1       |
|        | Operating System          | Linux / macOS / Windows                |         |
| Bottom | Hardware                  | Laptop, cluster, etc                   |         |

This scientific environment course will start at the bottom and work upwards. Clearly we cannot change the bottom two layers of the stack but there are advantages of keeping everyting above the operating system consistent, for example 

1. Installing your reserach software in larger scale computing resources
1. Transfering software between colleagues accurately 

Scientific software is moving towards this model, for example Xsuite, RFTrack, WarpX can all be installed via a package manager. Most of your day to day work will be in the top of the software stack but importantly upgrading the underlying stack is very important.
